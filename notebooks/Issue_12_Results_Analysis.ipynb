```python
import pandas as pd
import numpy as np
pd.set_option("display.max_rows", 1000)
```

```python
full_df = pd.read_csv("/home/paulo/Documents/GIT/BCI_MsC/experiments/issue_12_openbmi/results.csv")
full_df = full_df[~full_df.algorithm.isin(["pca", "whitening"])]
# full_df = pd.read_csv("/home/paulo/Documents/GIT/BCI_MsC/experiments/issue_19/results.csv")

print(full_df.shape)
subset = set(full_df.columns) - set(["run"])
full_df = full_df.drop_duplicates(subset=subset)
print(full_df.shape)
full_df.columns
```

```python
full_df.run.unique(), full_df.uid.unique()
```

```python
df = full_df.groupby(["uid", "classifier", "algorithm"], as_index=False).mean()
```

```python
df.query("(uid == 5)").sort_values(by="Kappa", ascending=False)
```

```python
sorted_all = df.groupby(["uid", "classifier", "algorithm"], as_index=False).mean().reset_index()#[["uid", "Kappa"]]
display(sorted_all.query("uid == 3").sort_values(by="Kappa").tail(1)[["uid", "classifier", "algorithm", "Kappa"]].round(3))
sorted_all["order"] = sorted_all.groupby("uid").rank(ascending=False)["Kappa"]
sorted_all.query("order == 1")[["uid", "classifier", "algorithm", "Kappa"]].round(3)
```

```python
sorted_alg = df.groupby(["uid", "algorithm"], as_index=False).agg([np.mean, np.std]).Kappa.sort_values(by=["uid", "mean"])
sorted_alg["order"] = sorted_alg.groupby("uid").rank(ascending=False)["mean"]
sorted_alg
```

```python
sorted_alg.query("order == 1").drop("order", axis=1).round(3)
```

```python
sorted_alg.groupby("algorithm").median()[["order"]].sort_values(by="order")
```

```python
sorted_clf = df.groupby(["uid", "classifier"], as_index=False).agg([np.mean, np.std]).Kappa.sort_values(by=["uid", "mean"])
sorted_clf["order"] = sorted_clf.groupby("uid").rank(ascending=False)["mean"]
sorted_clf
```

```python
sorted_clf.query("order == 1").drop("order", axis=1).round(3)
```

```python
sorted_clf.groupby("classifier").median()[["order"]].sort_values(by="order")
```
